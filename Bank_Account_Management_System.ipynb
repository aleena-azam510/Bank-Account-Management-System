Bank Account Management System

Write a Python program that simulates a simple bank account management system. The system should allow:

Creating an account for a user with a name and an initial balance.

Depositing money into the account.

Withdrawing money from the account (ensure the balance doesnâ€™t go negative).

Checking the account balance.

Transferring money between two accounts.

Use the following specifications:

Create a class `BankAccount` with methods to handle the actions listed above.

Ensure proper error handling (e.g., preventing overdraft, invalid input).

Use multiple objects of the `BankAccount` class to simulate multiple users.

Display all relevant information after each action.











```python
class BankAccount:
  def __init__(self,name,initial_balance):
    self.name = name
    self.balance = initial_balance
    print("....................................................................")


  def check_balance(self):
    print(f"Your account balance  is {self.balance}")
    if self.balance < 0:
      print("Insufficient balance")

  def deposit_money(self,amount):
    if amount > 0 :
      self.balance += amount
      print(f"{amount} rupees deposited successfully")
    else:
            print("Invalid deposit amount. Please enter a positive amount.")

  def withdraw_money(self,amount):
      if amount > 0 and amount <= self.balance:
         self.balance -= amount
         print(f"{amount}rupees withdraw successfully")
      else:
         print("Insufficient balance")
         print("You have not enough money to withdraw")

  def transferring_money(self,initial_balance,recipent_name,amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            recipient_account = BankAccount(recipent_name, initial_balance)  # Use recipent_name here
            recipient_account.balance += amount
            print(f"{amount} rupees transferred successfully to {recipient_name}!")
        else:
         print("Insufficient balance")
         print(f"You can not transfer the money")

print("Welcome to the Bank Account Management System")
print("-------------------------------------------------")
print(".................................................")
name = input("Enter your name: ")
initial_balance = int(input("Enter initial balance: "))
print(f"Hello {name}, Welcome to the Bank Account Management System")
print(f"{name} , your initial balance is {initial_balance}")
print("-------------------------------------------------")

account = BankAccount(name,initial_balance)

for i in range(6):
  print("What would you like to do?")
  print("1. Check Balance")
  print("2. Deposit Money")
  print("3. Withdraw Money")
  print("4. Transfer Money")
  print("5. Exit")
  print("-------------------------------------------------")
  choice = int(input("Enter your choice: "))
  print("-------------------------------------------------")
  if choice == 1:
    account.check_balance()
    print("-------------------------------------------------")
  elif choice == 2:
    amount = int(input("Enter deposit amount: "))
    account.deposit_money(amount)
    account.check_balance()
    print("-------------------------------------------------")
  elif choice == 3:
    amount = int(input("Enter withdraw amount: "))
    account.withdraw_money(amount)
    account.check_balance()
    print("-------------------------------------------------")
  elif choice == 4:
    recipient_name = input("Enter recipient name: ")
    initial_balance = 0
    amount = int(input("Enter transfer amount: "))
    account.transferring_money(initial_balance,recipient_name, amount)
    account.check_balance()
    print("-------------------------------------------------")
  elif choice == 5:
    print("Thank you for using the Bank Account Management System")
    print("-------------------------------------------------")
    break
  else:
    print("Invalid choice")
```
